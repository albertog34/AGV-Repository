```python
# Let's try to create a game in which the user has a total of 3 chances to guess a random number generated from 1 to 100.
# As long as the user does not guess the number, we will give him/her clues until he/she finally manages to guess it.

def game():

    # We start by welcoming the user:

    import colorama as col
    from colorama import Fore
    
    print(Fore.BLUE+"Welcome to the Guess or die game!")
    print("*"*50)
    print(Fore.BLUE+"Will you be able to guess a number from 1 to 100 from a maximum of 3 clues?")
    print("*"*50)
    print(Fore.BLUE+"Let's check it out!")
    print("*"*50)

    # Next, the first thing we will do is generate a random number that will be the user's target.

    from random import randint

    target_number=int(randint(1,100))

    # The next thing we will do is to start building the tracks that will be shown to the user. We will start with the first one
    # We'll start simple, we'll tell the user if the number is odd or even.

    if target_number%2==0:
        print("Clue1: The number we are looking for is even")
        print("*"*50)
    else:
        print("Clue 1: The number we are looking for is odd")
        print("*"*50)

    # We will then ask the user to enter the number they think it is, and compare it against our searched value. If it does not match, the second hint will be shown.

    user_number_1=int(input("What number do you think it is?"))
    print("*"*50)

    if user_number_1==target_number:
        print(Fore.GREEN+f"Congratulations! The number we were looking for was {target_number}!")
        print("*"*50)
    else:
        print(Fore.RED+"Wrong! It's difficult with just one clue... isn't it? Let's go for the second one!")
        print("*"*50)

        # Since the user has not guessed correctly, we will go for the second clue. This time, we'll make him think a bit harder and tell him all the numbers by which it is divisible.

        divisible_numbers=[]
        for i in range(1,target_number):
            if target_number%i==0:
                divisible_numbers.append(i)
            else:
                pass

        # Once this has been done, we show the user the results

        if len(divisible_numbers)>1:
            print(f"The number we are looking for is divisible by all these numbers: {divisible_numbers}")
            print("*"*50)
        else:
            print("The number we are looking for is prime!")
            print("*"*50)

        # We then ask the user again if he/she already knows the number we are looking for, and check it again against the initial number.

        user_number_2=int(input("Now, what number do you think it is?"))
        print("*"*50)
        if user_number_2==target_number:
            print(Fore.GREEN+f"Congratulations! The number we were looking for was {target_number}!")
            print("*"*50)
        else:
            print(Fore.RED+"You're getting close to bankruptcy, huh? Need a third clue to guess it? What a low IQ...")
            print("*"*50)

            # Since the user still can't guess, we give the last clue of all. On this occasion, we are going to narrow down a range of +-10 numbers between which the target is to be found.

            lower_number_interval=int(target_number*(1-(int(randint(0,10))/100)))
            higher_number_interval=int(target_number*(1+(int(randint(0,10))/100)))

            print(f"Come on, man, I'll make it easy for you: your number is between {lower_number_interval} and {higher_number_interval}, both inclusive")
            print("*"*50)

            user_number_3=int(input("Come on, mate, last chance, do you know what number we're talking about or not?"))
            print("*"*50)
            if user_number_3==target_number:
                print(Fore.GREEN+f"Hallelujah! The number we were looking for was {target_number}!")
                print("*"*50)
            else:
                print(Fore.RED+"Man, you're a damn loser.... Seriously, do something other than maths ;)")
                print("*"*50)
                print(Fore.RED+f"The number we were looking for was {target_number}!")
                print("*"*50)
                print(Fore.BLUE+"THE GAME IS END. TRY AGAIN!")

game()
```
