```python
# Creating a tennis scoreboard

import pandas as pd

# First of all, we define the individual scoreboard for each  

pair1=0
pair2=0

# Since there is no equidistance between the point values (+15,+15,+10), we perform an individual point count and map it to the possible padel / tennis point values.

possible_points=[0,1,2,3]
tennis_points=[0,15,30,40]
points_mapping=list(zip(possible_points,tennis_points))
dataframe=pd.DataFrame(points_mapping,columns=["possible_points","tennis_points"])

# Subsequently, we developed the code. As long as the point count is less than 4 in either pair, the user continues to be asked who has won the point
# In order to display the points, we will not show the "individual points" scored by each pair, but we will look for their equivalent in the "points_mapping" dataframe we have prepared.

while pair1<4 and pair2<4:
    point_winner=input("Who has scored the point?")
    if point_winner=="pair1":
        pair1+=1
    elif point_winner=="pair2":
        pair2+=1    
    if pair1<4 and pair2<4:
        print(int(dataframe.loc[dataframe["possible_points"]==pair1,"tennis_points"]),"-",int(dataframe.loc[dataframe["possible_points"]==pair2,"tennis_points"]))

# At this point, we define the most complicated part, the "deuce". To do this, we establish the difference in points between pair 1 and 2, and we establish that, if this is 2 (or more, which is obviously impossible), the game ends.
    
    if pair1==3 and pair2==3:
        difference=abs(pair1-pair2)
        while difference<=1:
            deuce_winner=input("Who has scored the point?")
            if deuce_winner=="pair1":
                print("Adv","-",40)
                pair1+=1
                adv_winner=input("Who has scored the point?")
                if deuce_winner==adv_winner:
                    break #If pair 1 have won the first and second point of the "deuce" situation, the game ends.
                else:
                    pair1-=1 #If pair 1 have lost the second point of the "deuce" situation, the the loop keeps repeating because the difference between pair 1 and 2 is no longer than 1
                    print(int(dataframe.loc[dataframe["possible_points"]==pair1,"tennis_points"]),"-",int(dataframe.loc[dataframe["possible_points"]==pair2,"tennis_points"]))
            elif deuce_winner=="pair2":
                print(40,"-","Adv")
                pair2+=1
                adv_winner=input("Who has scored the point?")
                if deuce_winner==adv_winner:
                    break #If pair 2 have won the first and second point of the "deuce" situation, the game ends.
                else:
                    pair2-=1 #If pair 2 have lost the second point of the "deuce" situation, the the loop keeps repeating because the difference between pair 1 and 2 is no longer than 1
                    print(int(dataframe.loc[dataframe["possible_points"]==pair1,"tennis_points"]),"-",int(dataframe.loc[dataframe["possible_points"]==pair2,"tennis_points"]))            

else:
    if pair1>3:    
        print("Game achieved for pair 1!")
    else:
        print("Game achieved for pair 2!")
```
